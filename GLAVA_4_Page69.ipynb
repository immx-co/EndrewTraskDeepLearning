```python
weight = 0.5
input = 0.5
goal_prediction = 0.8

step_amount = 0.001

for iteration in range(1500):
    prediction = input * weight
    error = (prediction - goal_prediction) ** 2
    
    
    up_prediction = input * (weight + step_amount)
    up_error = (up_prediction - goal_prediction) ** 2
    
    down_prediction = input * (weight - step_amount)
    down_error = (down_prediction - goal_prediction) ** 2
    
    print(f'Error: {error} Prediction: {prediction}')
    
    if down_error < up_error:
        weight = weight - step_amount
    elif down_error > up_error:
        weight = weight + step_amount
```

```python
weight = 0.5
goal_pred = 0.8
input = 0.5

for iteration in range(100):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    direction_and_amount = (pred - goal_pred) * input
    weight = weight - direction_and_amount
    
    print(f'Error: {error} Prediction: {pred} direction_and_amount: {direction_and_amount}')
```

```python
weight = 0.1

alpha = 0.01

def neural_network(input, weight):
    prediction = input * weight
    return prediction

number_of_toes = [8.5]
win_or_lose_binary = [1]

input = number_of_toes[0]
goal_pred = win_or_lose_binary[0]

pred = neural_network(input, weight)

error = (pred - goal_pred) ** 2

delta = pred - goal_pred

weight_delta = input * delta

weight -= weight_delta * alpha

print(error, delta, pred, weight_delta, weight)
```

```python
# обучение просто умеьншает ошибку

weight, goal_pred, input = 0.0, 0.8, 0.5

for iteration in range(25):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * input
    weight -= weight_delta
    print(f'Error: {error} Prediction: {pred}')
```

```python
# Обучение - это корректировка веса для уменьшения ошибки до 0

alpha = 0.0001
weight = 0.5
goal_pred = 0.8
input = 2

for iteration in range(500):
    pred = input * weight
    error = (pred - goal_pred) ** 2
#     delta = pred - goal_pred
#     weight_delta = delta * input
    derivative = input * (pred - goal_pred)
    weight = weight - (alpha * derivative)
    
    print(f'Error: {error} Prediction: {pred}')
```
