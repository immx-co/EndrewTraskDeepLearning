```python
weight = 5
input = 2
goal_prediction = 0.75
alpha = 0.01

for iteration in range(1000):
    pred = weight * input
    error = (pred - goal_prediction) ** 2
    delta = pred - goal_prediction
    weight = weight - alpha * (input * delta)
    
    print(f'Error: {error} Prediction: {pred}')
```
